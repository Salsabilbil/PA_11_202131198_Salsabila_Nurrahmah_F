```python
#Import Library 
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#to read rgb image
img = cv2.imread('sevatu.jpg')
```

```python
#to display rgb images
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
copy_img = img.copy()
plt.imshow(img)
```

```python
#now convert the RGB image to grayscale
lower = np.array([30,30,30])
higher = np.array([250,250,250])
```

```python
mask = cv2.inRange(img,lower,higher)
```

```python
#to create an image with a gray color
plt.imshow(mask,'gray')
```

```python
#detect the contours
cont,_ = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
#to create blank images
blank_image = np.zeros((1500,1500,3), np.uint8)
blank_image[:,0:1500//2] = (255,255,255)
blank_image[:,1500//2:1500] = (255,255,255)
```

```python
contour = cv2.drawContours(blank_image, cont, -1, 255, 0)
```

```python
#to create a blank image with a 180 degree rotation
contour = cv2.rotate(contour,cv2.ROTATE_180)
plt.imshow(contour)
```

```python
#to create draw  contours images
cont_img = cv2.drawContours(img,cont,-1,255,10)
```

```python
#to show contours images
plt.imshow(cont_img)
```

```python
c = max(cont,key = cv2.contourArea)
```

```python
x,y,w,h=cv2.boundingRect(c)
```

```python
#Visualize the data Structure
cv2.rectangle(img,(x,y),(x+w,y+h),(-1,255,5),0)
plt.imshow(img)
```

```python
cropped_image = copy_img[y:y+h,x:x+w]
```

```python
#to show cropped image
plt.imshow(cropped_image)
```

```python
#to display the three final images
plt.figure(figsize=(20,4))
plt.subplot(1,3,1),plt.imshow(copy_img)
plt.subplot(1,3,2),plt.imshow(contour)
plt.subplot(1,3,3),plt.imshow(cont_img)
```
